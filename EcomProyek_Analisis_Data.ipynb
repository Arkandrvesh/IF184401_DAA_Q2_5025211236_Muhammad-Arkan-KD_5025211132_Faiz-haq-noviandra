## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Merge data to get customer information for each order
merged_data = pd.merge(orders, customers, on='customer_id', how='left')

# Convert order_purchase_timestamp to datetime format
merged_data['order_purchase_timestamp'] = pd.to_datetime(merged_data['order_purchase_timestamp'])

# Extract year and month from order_purchase_timestamp
merged_data['year'] = merged_data['order_purchase_timestamp'].dt.year
merged_data['month'] = merged_data['order_purchase_timestamp'].dt.month
```

```python
cust_order_data = cust_order_data.duplicated().sum()
cust_order_data
```

```python
customers.info()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# jumlah pesanan berulang dari pelanggan aktif bulanan setiap tahun
monthly_active_customers = merged_data.groupby(['customer_unique_id', 'year', 'month']).size().reset_index(name='order_count')

# Group data by year to find the average number of repeat orders per month for each year
repeat_orders_per_year = monthly_active_customers.groupby('year')['order_count'].mean()

# Visualiisasi
plt.figure(figsize=(10, 6))
repeat_orders_per_year.plot(kind='bar', color='skyblue')
plt.title('Average Repeat Orders from Monthly Active Customers Each Year')
plt.xlabel('Year')
plt.ylabel('Average Repeat Orders')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
print("Average number of repeat orders from monthly active consumers each year:")
print(repeat_orders_per_year)
```
